Сделайте краткий обзор какой-нибудь научной работы посвященной тому или иному алгоритму для object detection, который не рассматривался на уроке. ### Будем рассматривать архитектуры семейства YOLOv5

Было рассмотрено семейство алгоритмов YOLOv5 (YOLOv5x, YOLOv5l, YOLOv5m, YOLOv5s)

### Плюсы
 1. Чрезвычайно просты в использовании для разработчика, внедряющего технологии компьютерного зрения в приложение.
 2. Встроенный механизм "умной" аугментации Mosaic augmentation, решающий проблему пропуска мелких объектов другими сетями.
 3. Быстрая скорость обучения по сравнению с аналогичными сетями YOLOv4.

### Минусы
 1. Нет научной статьи сопровождающий релиз данной архитектуры.
 2. Уступает YOLOv4 по многим важным параметрам.

### Трудности применения
 Одна из основных задач, поставленная автором данной архитетктуры, как я понял из одного интервью с ним, была как раз в том, чтобы максимально упростить использование данной сети для решения любых связанных задач. Жалоб на трудности применения в комментариях к статьям и на GitHub не обнаружил.

### Критика алгоритма в профессиональном сообществе
Автор данного алгоритма подвергся критике со стороны разработчика YOLOv4 и других специалистов области машинного обучения.
Основная причина связана с выбором имени для алгоритма, подразумевающем развитие сети YOLOv4.
На деле же, этот алгоритм скорее находится рядом c YOLOv4 и не имеет черт YOLOv4, которых нет у YOLOv3.
То есть это по сути другая ветвь развития YOLOv3, которая больше ориентирована на скорость работы и простоту, но по точности уступает YOLOv4.
Также автор не выпустил вместе с релизом научную работу, описывающую подробно все его доработки.
А также сравнения YOLOv5 автором с YOLOv4 были непрезентабельны из-за разных датасетов и условий проведения тестов.
